```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
df = pd.read_csv("enrollments_data.csv")
```

# Enrollments by degree

```python
# see total number of enrollments for each year

df_1 = df[
    (df['ISCED'] == 'Total') &
    (df['Status of student'] == 'Total') &
    (df['Gender'] == 'Total') &
    (df['Institution type'] == 'Total') &
    (df['Registration status'] == 'Total') &
    (df['Field of study'] == 'Total') &
    (df['Location'] == 'Canada')
]

df_2 = df[
    (df['ISCED'] == "Bachelor's or equivalent") &
    (df['Status of student'] == 'Total') &
    (df['Gender'] == 'Total') &
    (df['Institution type'] == 'Total') &
    (df['Registration status'] == 'Total') &
    (df['Field of study'] == 'Total') &
    (df['Location'] == 'Canada')
]

df_3 = df[
    (df['ISCED'] == "Master's or equivalent") &
    (df['Status of student'] == 'Total') &
    (df['Gender'] == 'Total') &
    (df['Institution type'] == 'Total') &
    (df['Registration status'] == 'Total') &
    (df['Field of study'] == 'Total') &
    (df['Location'] == 'Canada')
]

df_4 = df[
    (df['ISCED'] == "Doctoral or equivalent") &
    (df['Status of student'] == 'Total') &
    (df['Gender'] == 'Total') &
    (df['Institution type'] == 'Total') &
    (df['Registration status'] == 'Total') &
    (df['Field of study'] == 'Total') &
    (df['Location'] == 'Canada')
]

# Align indices for all DataFrames
df_1 = df_1.reset_index(drop=True)
df_2 = df_2.reset_index(drop=True)
df_3 = df_3.reset_index(drop=True)
df_4 = df_4.reset_index(drop=True)
```

```python
# Create the figure and axis
plt.figure(figsize=(10, 6))  # Set the figure size

# Plot each dataset with distinct colors and labels
plt.plot(df_1['Date'], df_1['Value'], marker='o', linestyle='-', color='b', label='Total')
plt.plot(df_2['Date'], df_2['Value'], marker='o', linestyle='-', color='y', label="Bachelor")
plt.plot(df_3['Date'], df_3['Value'], marker='o', linestyle='-', color='g', label='Master')
plt.plot(df_4['Date'], df_4['Value'], marker='o', linestyle='-', color='r', label='Doctor')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Students')
plt.title('Line Chart of Students Over Years')

# Add a legend to differentiate the datasets
plt.legend(loc='upper left')  # You can adjust the location

# Optionally, add grid lines for better readability
plt.grid(True)

# Display the plot
plt.show()
```

```python
# Create the figure and axis
plt.figure(figsize=(10, 6))  # Set the figure size

# Plot each dataset with distinct colors and labels
plt.plot(df_2['Date'], df_2['Value']/df_1['Value'], marker='o', linestyle='-', color='y', label="Bachelor")
plt.plot(df_3['Date'], df_3['Value']/df_1['Value'], marker='o', linestyle='-', color='g', label='Master')
plt.plot(df_4['Date'], df_4['Value']/df_1['Value'], marker='o', linestyle='-', color='r', label='Doctor')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Students')
plt.title('Line Chart of Students Over Years')

# Add a legend to differentiate the datasets
plt.legend(loc='upper left')  # You can adjust the location

# Optionally, add grid lines for better readability
plt.grid(True)

# Display the plot
plt.show()
```

# International students

```python
# See total number of enrollments for each year

df_1 = df[
    (df['ISCED'] == 'Total') &
    (df['Status of student'] == 'Total') &
    (df['Gender'] == 'Total') &
    (df['Institution type'] == 'Total') &
    (df['Registration status'] == 'Total') &
    (df['Field of study'] == 'Total') &
    (df['Location'] == 'Canada')
].copy()  # Ensure a copy is made

df_2 = df[
    (df['ISCED'] == 'Total') &
    (df['Status of student'] == 'International students') &
    (df['Gender'] == 'Total') &
    (df['Institution type'] == 'Total') &
    (df['Registration status'] == 'Total') &
    (df['Field of study'] == 'Total') &
    (df['Location'] == 'Canada')
].copy()  # Ensure a copy is made
df_2.loc[:, 'Revenue'] = df_2['Value'] * 3  # Use .loc[] for assignment

df_3 = df[
    (df['ISCED'] == 'Total') &
    (df['Status of student'] == 'Canadian students') &
    (df['Gender'] == 'Total') &
    (df['Institution type'] == 'Total') &
    (df['Registration status'] == 'Total') &
    (df['Field of study'] == 'Total') &
    (df['Location'] == 'Canada')
].copy()  # Ensure a copy is made
df_3.loc[:, 'Revenue'] = df_3['Value'] * 1  # Use .loc[] for assignment

# Align indices for all DataFrames
df_1 = df_1.reset_index(drop=True)
df_2 = df_2.reset_index(drop=True)
df_3 = df_3.reset_index(drop=True)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming 'Date' column is your x-axis and 'Value' columns are your y-axis
x = np.arange(len(df_2['Date']))  # Create an array of positions for the x-axis

# Create the figure and axis
plt.figure(figsize=(10, 6))  # Set the figure size

# Plot the stacked bar chart without 'Total'
plt.bar(x, df_2['Value'], color='g', label='International')
plt.bar(x, df_3['Value'], color='y', bottom=df_2['Value'], label='Canadian')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Students, mln')
plt.title('Amount of International and Canadian Students 1992 - 2021')

# Add a legend to differentiate the datasets
plt.legend(loc='upper left')  # You can adjust the location

# Set x-ticks with the actual dates
plt.xticks(x, df_2['Date'], rotation=45)

# Optionally, add grid lines for better readability
plt.grid(True, axis = 'y')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming 'Date' column is your x-axis and 'Value' columns are your y-axis
x = np.arange(len(df_2['Date']))  # Create an array of positions for the x-axis

# Create the figure and axis
plt.figure(figsize=(10, 6))  # Set the figure size

# Plot the stacked bar chart without 'Total'
plt.bar(x, df_2['Revenue'], color='g', label='International')
plt.bar(x, df_3['Revenue'], color='y', bottom=df_2['Revenue'], label='Canadian')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Arbitrary units')
plt.title('Contribution to universities revenue from International and Canadian students')

# Add a legend to differentiate the datasets
plt.legend(loc='upper left')  # You can adjust the location

# Set x-ticks with the actual dates
plt.xticks(x, df_2['Date'], rotation=45)

# Optionally, add grid lines for better readability
plt.grid(True, axis = 'y')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Calculate the total revenue for each year
total_revenue = df_2['Revenue'] + df_3['Revenue']

# Calculate the percentage for each category
df_2_percentage = df_2['Revenue'] / total_revenue * 100
df_3_percentage = df_3['Revenue'] / total_revenue * 100

# Assuming 'Date' column is your x-axis and 'Value' columns are your y-axis
x = np.arange(len(df_2['Date']))  # Create an array of positions for the x-axis

# Create the figure and axis
plt.figure(figsize=(10, 6))  # Set the figure size

# Plot the percentage stacked bar chart without 'Total'
plt.bar(x, df_2_percentage, color='g', label='International')
plt.bar(x, df_3_percentage, color='y', bottom=df_2_percentage, label='Canadian')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Revenue contribution, %')
plt.title('Relative contribution to universities revenue from International and Canadian students')

# Add a legend to differentiate the datasets
plt.legend(loc='upper left')  # You can adjust the location

# Set x-ticks with the actual dates
plt.xticks(x, df_2['Date'], rotation=45)

# Optionally, add grid lines for better readability
plt.grid(True, axis = 'y')

# Display the plot
plt.show()
```
